```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Applying threshold (separate the background and the foreground objects)
thresh = cv2.threshold(gray, 0, 255,cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
plt.imshow(gray)
```

```python
# apply connected component analysis to the thresholded image
output = cv2.connectedComponentsWithStats(thresh, 8, cv2.CV_32S)
(numLabels, labels, stats, centroids) = output
```

## CCA via SKimage Libarary

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio.v3 as iio
import skimage.color
import skimage.filters
import skimage.measure
```

- Otsu's method is a automatic image thresholding method:
The method aims to find an optimal threshold value to separate an image into two classes, typically foreground and background, based on the grayscale intensity values of its pixels.

```python
def connected_components(filename, ostu_th, sigma=1.0, t=0.5, connectivity=2):
    # load the image
    image = iio.imread(filename)
    # convert the image to grayscale
    gray_image = skimage.color.rgb2gray(image)
    # denoise the image with a Gaussian filter
    blurred_image = skimage.filters.gaussian(gray_image, sigma=sigma)
    # mask the image according to threshold
    binary_mask = blurred_image < t
    # perform connected component analysis
    labeled_image, count = skimage.measure.label(ostu_th,
                                                 connectivity=connectivity, return_num=True)
    return labeled_image, count
```

```python
labeled_image, count = connected_components(datapath, thresh, sigma=2.0, t=0.9, connectivity=2)

print("Number of pores: " + str(count))
fig, ax = plt.subplots()
plt.imshow(labeled_image)
plt.axis("off");
```
